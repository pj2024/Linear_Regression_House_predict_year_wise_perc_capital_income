```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.figure(figsize=(6,5))
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
import warnings
warnings.filterwarnings('ignore')
reg.predict([[3300]])
```

```python
plt.figure(figsize=(6,5))
plt.xlabel('Area(sqr ft)')
plt.ylabel('Price(US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
plt.show()
```

```python
home_price_df = pd.read_csv('canada_per_capita_income.csv')
home_price_df
```

```python
plt.figure(figsize=(6,5))
plt.xlabel('Year')
plt.ylabel('Per Capital Income(US$)')
plt.scatter(home_price_df['year'],home_price_df['per capita income (US$)'], color='red', marker='+')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
lr.fit(home_price_df[['year']], home_price_df['per capita income (US$)'])
```

```python
lr.predict([[2020]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# y = m*x+c
828.46507522*2020+(-1632210.7578554575)
```

```python
plt.figure(figsize=(6,5))
plt.xlabel('Year')
plt.ylabel('Per Capital Income(US$)')
plt.scatter(home_price_df['year'],home_price_df['per capita income (US$)'], color='red', marker='+')
plt.plot(home_price_df['year'], lr.predict(home_price_df[['year']]), color='blue')
plt.show()
```
